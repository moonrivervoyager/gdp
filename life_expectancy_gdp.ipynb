```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
df = pd.read_csv("all_data.csv")  # Ensure correct file path

# Normalize GDP for better comparison (convert to trillions)
df["GDP (Trillions)"] = df["GDP"] / 1e12  

# FIX: Ensure only one entry per country (take the most recent year or average values)
df_grouped = df.groupby("Country", as_index=False).agg({"GDP (Trillions)": "mean", "Life expectancy at birth (years)": "mean"})

# Set figure size
fig, ax1 = plt.subplots(figsize=(12, 6))

# Width for bars
bar_width = 0.4  

# Get positions for bars
x_positions = range(len(df_grouped["Country"]))

# Bar plot for GDP (left y-axis)
ax1.bar(x_positions, df_grouped["GDP (Trillions)"], width=bar_width, color="royalblue", alpha=0.7, label="GDP (Trillions)")

# Create a second y-axis for Life Expectancy
ax2 = ax1.twinx()
ax2.bar([x + bar_width for x in x_positions], df_grouped["Life expectancy at birth (years)"], width=bar_width, color="darkorange", alpha=0.7, label="Life Expectancy")

# Set x-ticks and labels
ax1.set_xticks([x + bar_width / 2 for x in x_positions])  # Center labels
ax1.set_xticklabels(df_grouped["Country"], rotation=45, ha="right")  # Rotate for better readability

# Adjust y-axis limits
ax1.set_ylim(0, df_grouped["GDP (Trillions)"].max() * 1.2)  # Scale GDP properly
ax2.set_ylim(40, 90)  # Set a reasonable range for Life Expectancy

# Improve labels
ax1.set_xlabel("Country", fontsize=12)
ax1.set_ylabel("GDP (Trillions USD)", fontsize=12, color="royalblue")
ax2.set_ylabel("Life Expectancy (Years)", fontsize=12, color="darkorange")

# Improve title
plt.title("GDP and Life Expectancy by Country", fontsize=14)

# Fix legends
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
ax1.set_xticks([x + bar_width / 2 for x in x_positions])
ax1.set_xticklabels(df_grouped["Country"], rotation=30, ha="right")


# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
df = pd.read_csv("all_data.csv")  # Ensure correct file path

# Normalize GDP for better comparison (convert to trillions)
df["GDP (Trillions)"] = df["GDP"] / 1e12  

# Set figure size
fig, ax1 = plt.subplots(figsize=(14, 6))

# Custom color palettes (darker colors for better readability)
gdp_palette = sns.color_palette("dark:blue", n_colors=len(df["Country"].unique()))
life_palette = sns.color_palette("dark:orange", n_colors=len(df["Country"].unique()))

# Line plot for GDP (left y-axis)
gdp_plot = sns.lineplot(
    data=df, x="Year", y="GDP (Trillions)", hue="Country",
    marker="o", linewidth=2, ax=ax1, palette=gdp_palette
)
ax1.set_ylabel("GDP (Trillions USD)", fontsize=12, color="royalblue")
ax1.set_xlabel("Year", fontsize=12)

# Create a second y-axis for Life Expectancy
ax2 = ax1.twinx()
life_plot = sns.lineplot(
    data=df, x="Year", y="Life expectancy at birth (years)", hue="Country",
    marker="s", linestyle="dashed", linewidth=2, ax=ax2, palette=life_palette
)
ax2.set_ylabel("Life Expectancy (Years)", fontsize=12, color="darkorange")

# Improve title and subtitle for better storytelling
plt.title("GDP & Life Expectancy Trends Over Time", fontsize=14)
plt.suptitle("Analyzing Economic Growth & Health Improvements (2000-2015)", fontsize=12, y=1.02)

# Adjust legend placement **completely outside** the plot
gdp_legend = ax1.legend(title="GDP (Trillions)", loc="upper left", bbox_to_anchor=(-0.2, 1.1))  
life_legend = ax2.legend(title="Life Expectancy", loc="upper right", bbox_to_anchor=(1.3, 1.1))


# Match legend markers to actual line styles (solid for GDP, dashed for Life Expectancy)
for line in gdp_legend.get_lines():
    line.set_linestyle('-')
for line in life_legend.get_lines():
    line.set_linestyle('--')

# Add gridlines for improved readability
ax1.grid(True, linestyle="--", alpha=0.5)

# Increase spacing to avoid legend crowding
plt.subplots_adjust(top=0.85)

# Show the plot
plt.show()
```
